```python
def calculatePositValue(binary, posit, expo):
    sign = 0
    regimesign = 0
    isRegime = True
    runlength = 0
    expo_bitvalue = 0
    fraction_bitvalue = 0
    fractionlength = 0
    i = int(binary, 2)
    INF = "1"
    j = len(binary) - 1
    
    while(j > 0):
        INF = INF + "0"
        j -= 1
    
    
    if binary == INF:
        value = "INF"

    elif i == 0:
        value = 0

    else:
        for j in range(len(binary)):
            if (j == 0):
                sign=int(binary[j])
                if sign == 0:
                    sign=1

                elif sign == 1:
                    sign=-1
                    # Two's complements
                    binary=dec2bin1(-i)[32-posit:32]
                    while(len(binary) < posit):
                        binary="0" + binary

            elif (j == 1):
                
                regimesign=int(binary[j])
                runlength += 1

            else:
                if (isRegime):
                    if(int(binary[j]) == regimesign):
                        runlength += 1

                    else:
                        isRegime=False
                        expo_start=runlength + 2
                        expoLength_max=expo_start + int(expo)
                        if (int(expo) > 0 and (expo_start) < len(binary)):
                            tmp_expo=binary[expo_start: len(binary)]

                            if (len(tmp_expo) > expo):
                                tmp_expo=tmp_expo[0: expo]


                            elif (len(tmp_expo) < expo):
                                missingzero=expo-len(tmp_expo)
                                while(missingzero > 0):
                                    tmp_expo += "0"
                                    missingzero -= 1


                            expo_bitvalue=int(tmp_expo, 2)


                        # Check fraction
                        if(expoLength_max < len(binary)):
                            fractionlength=len(binary) - expoLength_max
                            tmp_frac=binary[expoLength_max:len(binary)]
                            fraction_bitvalue=int(tmp_frac, 2)


                        break





        k=0

        if (regimesign == 0):
            k=-(runlength)

        else:
            k=runlength-1

        regime=pow(pow(2, pow(2, expo)), k)
        expo_value=pow(2, expo_bitvalue)
        fraction_max=pow(2, fractionlength)
        fraction_value=1 + (int(fraction_bitvalue)/fraction_max)
        value=sign * regime * expo_value * fraction_value



    return value
```

```python
expvalue = {
    "0":  1,
    "1": 2,
    "2": 4  ,
    "3": 8 , 
    "4":16,
    "5": 32
} 


def convertDToP(decimal,places , es):

    positMax = {}
    positMax.update({str(places) + '_'+str(es):(2**2**es)**(places-2)})
    positkeys = list(positMax.keys())
    
    for i in range(len(positkeys)):	

        positInfo = positkeys[i].split("_")
        ps = int(positInfo[0])
        es = int(positInfo[1])

        #Within range
        p_max = positMax[positkeys[i]]
        useed = pow(2, expvalue[str(es)])
        usigned_decimal = abs(decimal)
        sign = sign_check(decimal)
        posit = ""
        temp = usigned_decimal
        value = 0

        # handling so that there is no overflow
        if (usigned_decimal > p_max):

            if(sign<0):
                posit = "1"
                k = ps -2
                while(k>0):
                    posit += "0"
                    k -= 1

                posit += "1"

            else:
                posit = "0"
                k = ps-1
                while(k>0):
                    posit += "1"
                    k -= 1

        #Check for equal to 0
        elif (usigned_decimal==0):

            k = ps
            while(k>0):
                posit += "0"
                k -= 1

        #Check num is between 0 and 1
        elif (usigned_decimal < 1):

            #regime<1 => k<0

            posit += "0"
            while(True):
                temp2=temp*useed
                if (temp2>=useed):
                    break

                else:
                    temp = temp2
                    posit += "0"


            posit += "1"

        #Condtion bw 1 and useed satisfied
        elif (usigned_decimal >=1  and usigned_decimal < useed):
            #regime = 1 => k=1 => regime is 01 or 10
            posit = "010"
       
        #Regime Calculation
        elif (usigned_decimal>=useed ):
            #regime > 1
            j=0
            posit +="01"
            while (temp>=useed):
                temp = temp/useed
                j += 1
                posit += "1"

            posit +="0"

    

        if(usigned_decimal>0  and usigned_decimal <= p_max):

            #value is in exponential and fraction
            #Exp Calculation
            j=0
            while (temp>=2):
                temp = temp/2
                j += 1


            if(es>0):
                e_binary = dec2bin1(j)
                while (len(e_binary)<es):
                    e_binary = "0"+e_binary

                posit += e_binary


            f_fraction=dec2bin(temp)

            index = f_fraction.find('1') + 1
            posit += f_fraction[index:len(f_fraction)].strip(".")

            while (len(posit)< ps):
                posit += "0"

            if (len(posit)>ps):
                
                #Need to round faster method
                #Condition for no of iterations for rounding
                roundPlusOne = False
                if (posit[ps] == "1"):
                    n = ps+1
                    while(n<len(posit)):
                        if (posit[n] == "1"):
                            roundPlusOne = True
                            break
                        n += 1
                    if (not(roundPlusOne)):
                        #round to nearest even.
                        if (posit[ps-1]=="1"):
                            roundPlusOne = True



                if(roundPlusOne):
                    posit = posit[0:ps]
                    posit2 = int(posit, 2) +1
                    posit2=  dec2bin(posit2)
                    while(len(posit2)<len(posit)):
                        posit2 = "0"+posit2
                    
                    value2 = calculatePositValue(posit2, ps, es)
                    if (value2 == "INF"):
                        #Prevent overflow
                        value1 = calculatePositValue(posit, ps, es)
                        value = value1

                    else:
                        posit = posit2
                        value = value2


                else:

                    posit = posit[0:ps]
                    value1 = calculatePositValue(posit, ps, es)
                    value = value1

                    #prevent underflow
                    if (value==0):
                        posit2 = int(posit, 2) +1
                        posit2=  dec2bin(posit2)
                        while(len(posit2)<len(posit)):
                            posit2 = "0"+posit2

                        value2 = calculatePositValue(posit2, ps, es)
                        posit = posit2
                        value = value2




            if(sign<0):
                twoComplement = int(posit, 2)
                posit = dec2bin1(-twoComplement)[32-ps:32]
                value = -value	


        ##Value Calculation
        # if (value ==0):
        #     if(usigned_decimal==0):
        #         value=0
        #     else:
        #         value = calculatePositValue(posit, ps, es)
                
        ##Error Calculation       
        # r_error = 0
        # abs_error = 0
        # r_error = abs((decimal-value)/decimal)
        # abs_error = (decimal - value)

        msg = posit
        
        

    return msg
```

```python
def comp1Bit(b):
    return '1' if b=='0' else '0'
def rightShift(bStr):
    return bin(int(bStr,2)//4)[2:].rjust(len(bStr),'0') #always right shift 2 places
def sigmoid(num):
    return (1/(1+math.exp(-num)))
```

```python
def sigmoidApprox(num,n):
    pStr = convertDToP(num,n,0)
    pstr = comp1Bit(pStr[0])+pStr[1:]
    return calculatePositValue(rightShift(pstr),n,0)

def tanhApprox(num,n):
    return(2*sigmoidApprox(2*num,n)-1)
```

```python
def Plot(r,n):
    fl=list(np.append(np.linspace(-r,0,2*r),(np.linspace(0.6,r,2*r))))
    plt.subplot(1,2,1)
    plt.plot(fl,[sigmoid(i) for i in fl],color="orange")
    plt.plot(fl,[sigmoidApprox(i,n) for i in fl])
    plt.legend(["approx sigmoid","exact sigmoid"])
    plt.subplot(1,2,2)
    plt.plot(fl,[sigmoid(i)-sigmoidApprox(i,n) for i in fl])
    plt.title("error")
    plt.show()
    plt.subplot(1,2,1)
    plt.plot(fl,[tanhApprox(i,n) for i in fl])
    plt.plot(fl,[math.tanh(i) for i in fl])
    plt.legend(["approx tanh","exact tanh"])
    plt.subplot(1,2,2)
    plt.plot(fl,[math.tanh(i)-tanhApprox(i,n) for i in fl])
    plt.title("error")
    plt.show()
```
