```python
def calculatePositValue(binary, posit, expo):
    sign = 0
    regimesign = 0
    isRegime = True
    runlength = 0
    expo_bitvalue = 0
    fraction_bitvalue = 0
    fractionlength = 0
    i = int(binary, 2)
    INF = "1"
    j = len(binary) - 1
    
    while(j > 0):
        INF = INF + "0"
        j -= 1
    
    
    if binary == INF:
        value = "INF"

    elif i == 0:
        value = 0

    else:
        for j in range(len(binary)):
            if (j == 0):
                sign=int(binary[j])
                if sign == 0:
                    sign=1

                elif sign == 1:
                    sign=-1
                    # Two's complements
                    binary=dec2bin1(-i)[32-posit:32]
                    while(len(binary) < posit):
                        binary="0" + binary

            elif (j == 1):
                
                regimesign=int(binary[j])
                runlength += 1

            else:
                if (isRegime):
                    if(int(binary[j]) == regimesign):
                        runlength += 1

                    else:
                        isRegime=False
                        expo_start=runlength + 2
                        expoLength_max=expo_start + int(expo)
                        if (int(expo) > 0 and (expo_start) < len(binary)):
                            tmp_expo=binary[expo_start: len(binary)]

                            if (len(tmp_expo) > expo):
                                tmp_expo=tmp_expo[0: expo]


                            elif (len(tmp_expo) < expo):
                                missingzero=expo-len(tmp_expo)
                                while(missingzero > 0):
                                    tmp_expo += "0"
                                    missingzero -= 1


                            expo_bitvalue=int(tmp_expo, 2)


                        # Check fraction
                        if(expoLength_max < len(binary)):
                            fractionlength=len(binary) - expoLength_max
                            tmp_frac=binary[expoLength_max:len(binary)]
                            fraction_bitvalue=int(tmp_frac, 2)


                        break





        k=0

        if (regimesign == 0):
            k=-(runlength)

        else:
            k=runlength-1

        regime=pow(pow(2, pow(2, expo)), k)
        expo_value=pow(2, expo_bitvalue)
        fraction_max=pow(2, fractionlength)
        fraction_value=1 + (int(fraction_bitvalue)/fraction_max)
        value=sign * regime * expo_value * fraction_value



    return value
```

```python
def genSigmoid(n,es):
    intNum = []
    twosCompInt = []
    positVal = []
    newPositVal = []
    for i in range(-pow(2,n-1),pow(2,n-1)):
        intNum.append(i)
        twosCompInt.append(bin(i+pow(2,n))[3:].rjust(n-1,'0').rjust(n,'1') if(bin(i)[0]=='-') else bin(i)[2:].rjust(n,'0'))
        positVal.append(calculatePositValue(twosCompInt[i+pow(2,n-1)],n,es))
    for i in twosCompInt:
        if(i!='INF'):
            t=(('1' if(i[0]=='0') else '0')+i[1:])[0:len(i)-1].rjust(n,'0')
            newPositVal.append(calculatePositValue(t,n,es))
    return (newPositVal,positVal,intNum)   
        
```
